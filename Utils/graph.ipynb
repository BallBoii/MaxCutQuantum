```python
import graph_tools
import networkx as nx
import numpy as np

# Create a random graph using the extended method.
G = nx.Graph.get_random_graph(10, prob=0.5, weighted=True, seed=42)

# Draw the graph using the new instance method.
G.draw_graph()

# Generate a random bitstring for the partition view.
bitstring = list(np.random.randint(0, 2, size=G.number_of_nodes()))

# Draw the bipartite partition of the graph.
G.draw_partition_graph(bitstring)
```
